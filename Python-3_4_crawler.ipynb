```python
#W5 #爬蟲 
import requests #取得網頁 HTML 原始碼
url = 'https://search.books.com.tw/search/query/key/python/cat/all'
r = requests.get(url) #r object(裡面有很多屬性eg. status_code))
print(r.status_code) # 200: success #100: 正在處理 #300:redirection #400: client error #500: server error

if r.status_code == requests.codes.ok: # requests library的code集合中的OK之變數（which is 200）
  print("OK!") 
# print(r.text)
```

```python
#beautifulsoup 函式庫: html的解析器（parser）
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser') #不一定要取名soup也沒關係
# print(soup.prettify()) #美化 # try it(也可不用)

title_tag = soup.title
print(title_tag)
# <title>博客來-目前您搜尋的關鍵字為: python</title>
print(title_tag.get_text()) #網頁的title 
print(title_tag.string)
# # 博客來-目前您搜尋的關鍵字為: python
```

```python
#soup.find_all('a') 可以取得所有 <a> 標籤（超連結）及其內容
a_tags = soup.find_all('a')  
for tag in a_tags:
  print(tag)
  print("======")
```

```python
#tag.get_text() 是個字串，為該標籤所包含的內容文字
a_tags = soup.find_all('a') 
#a_tags = soup.find_all('b') : 找所有粗體字
for tag in a_tags:
   print(tag.get_text()) #進一步把不需要的東西去掉
   print("======")
```

```python
#找價格所在
attr = {"class": "price clearfix"} 
# attr = {"class": "list-nav clearfix"} #dictionary 
price_tags = soup.find_all("ul", attrs = attr) #找屬性是class, 值是price的東西

for tag in price_tags:
  print(tag)
  print("======")
print(len(price_tags))
```

```python
# attr = {"class": "list-nav clearfix"}
attr = {"class": "price clearfix"} 
price_tags = soup.find_all("ul", attrs = attr)
#在 find_all() 中根據指定的屬性與值做篩選：
for tag in price_tags:
  print(tag.get_text().strip())
  print("======")

print(len(price_tags)) #一頁有24本書
```

```python
# attr = {"class": "list-nav clearfix"}
attr = {"class": "price clearfix"} 
price_tags = soup.find_all("ul", attrs = attr)
prices = []
for tag in price_tags:
  t = tag.get_text().strip()
  index_comma = t.find(",") #找‘,’和‘元’（因價格在這些元素附近））
  index_dollar = t.find("元")
  index_colon = t.find(":")
#自己觀察找規則撰寫：
  if index_comma >= 0:
     price_str = t[index_comma + 2:index_dollar - 1] #有些價格在, 後
  else:
     #有些價格在:後
     price_str = t[index_colon + 2:index_dollar - 1]
  prices.append(int(price_str))
  

print(prices)
print(len(prices)) # 一頁有24本書
print(sum(prices) / len(prices)) #平均價格
```

```python
#取得書名與產品頁面連結
# attr = {"rel": "mid_name"}
# attr = {"class": "box"}
attr = {"target": "_blank"}
title_tags = soup.find_all("a", attrs = attr)#找滿足此屬性的 a tag

titles = []
links = []

for tag in title_tags:
    # print(tag) # try it by yourself    
    title_str = tag.get_text().strip()
    print(title_str)
    titles.append(title_str)
    link_str = "https:" + tag["href"] #找"href"此屬性裡的值（網址）
    print(link_str)
    links.append(link_str)
    # print("===")
# print(titles)
# print(links)
```

```python
titles_a=[]
for i in titles:
    if i !='':
        titles_a.append(i)
print(titles_a)
titles_a=titles_a[:-8]
# print(len(titles_a))
# print(len(titles_a))
links_a=[]
for l in range(0, len(links), 2):
        links_a.append(links[l])
print(links_a)
links_a=links_a[:-1]
links_a=links_a[:-8]
# print(len(links_a))
```
